```python
import nltk
from jedi.inference import analysis
from textblob.en import polarity

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
```

```python
from nltk.tokenize import word_tokenize
from textblob import TextBlob
```

```python
#text = input("Enter the movie review")
text = """The movie is very interesting and engaging.
It has strong characters and a good storyline.
Some parts are slow but overall it is enjoyable."""

print("text is ", text)
```

```python
# Tokenize the text
tokens = word_tokenize(text)
print("tokens is ", tokens)
```

```python
# Remove Stoping words
from nltk.corpus import stopwords
# stopwords contains a set of stopping words and identified and removed from the tokens.
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in tokens if w.lower() not in stop_words]
print(filtered_words)
```

```python
# Lemmatization - Convert into simple words
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w, pos='v') for w in filtered_words]
print(lemmatized_words)
```

```python
# Stemming - Remove ed, ing, es etc. >> Lemmatizer will remove the stemming part
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = [stemmer.stem(w) for w in lemmatized_words]
print(stemmed_words)
```

```python
text = " ".join(lemmatized_words)   # convert list to string - Accept only text
analysis = TextBlob(text)
polarity = analysis.sentiment.polarity
print(polarity)
```

```python
if polarity > 0:
    print("GOOD REVIEW 😊")
elif polarity < 0:
    print("BAD REVIEW 😞")
else:
    print("NEUTRAL REVIEW 😐")
```

```python
# Sumarize the text and return high polarity sentence using TextBlob
blob = TextBlob(text) # Should pass text, not tokens
summary = max(blob.sentences, key=lambda s: s.sentiment.polarity)
print("Summary:", summary)
```
